```python
for url in URL[100:]:
    time.sleep(3)
#     try:
    driver.get(url)
    time.sleep(3)
#     except WebDriverException:
#         try:
#             time.sleep(3)
#             driver.get(url)
#         except WebDriverException:
#             time.sleep(3)
#             driver.get(url)
    
#     WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//span[@class="font-weight-bold"]')))
#Brand Name    
    try:
        brand=driver.find_element(By.XPATH,'//h1[@class="product-name hidden-sm-down"]')
        #for b in brand:
        Brand_Name.append(brand.text.split('-')[0].strip())
              
    except NoSuchElementException as e:
        Brand_Name.append("-")
        
#ProdName                                            
    try:
        pname=driver.find_elements(By.XPATH,'//h1[@class="product-name hidden-sm-down"]')
        pname1 = driver.find_elements(By.XPATH,'//h1[@class="product-name"]')
        if len(pname) >= 1:
            pname=driver.find_element(By.XPATH,'//h1[@class="product-name hidden-sm-down"]')
            ProdName.append(pname.text.strip())
        elif len(pname1) >= 1:
                pname1=driver.find_element(By.XPATH,'//h1[@class="product-name"]')
                ProdName.append(pname1.text.strip())
        else:    
            ProdName.append("-")
        
    except NoSuchElementException as e:
        ProdName.append("-")    
#Price
    try:
        price=driver.find_elements(By.XPATH,'//span[@class="sales"]/span')
        if len(price) >= 1:
            price=driver.find_element(By.XPATH,'//span[@class="sales"]/span')
            Price.append(price.text.strip())
        else:
            Price.append("-")
        
    except NoSuchElementException as e:
        Price.append("-")
      
    d = driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/th')
#Manufacturer                                           
    try:
        for i in range(0,len(d)):
            if driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/th')[i].text =='Manufacturer:':
                varient = driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/td')[i]
        #for b in varient:
        Manufacturer.append(varient.text)  

    except (NoSuchElementException, StaleElementReferenceException) as e:
        Manufacturer.append("-")
#Produt Type
    try:
        for i in range(0,len(d)):
            if driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/th')[i].text =='Product Type:':
                prob = driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/td')[i]
        #for b in prob:
        Produt_Type.append(prob.text)  
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Produt_Type.append("-")
        
# #Product Weight
    try:
        for i in range(0,len(d)):
            if driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/th')[i].text =='Product Weight:':
                prob_w = driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/td')[i]
#                 Produt_Weight.append(prob_w.text)
#             else:
#                 prob_w = '-'
#         for b in prob_w:
        Produt_Weight.append(prob_w.text)
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Produt_Weight.append("-")


#Item_Form
    try:
        for i in range(0,len(d)):
            if driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/th')[i].text =='Chemical Type:':
                item_form = driver.find_elements(By.XPATH,'//table[@class="table table-striped table-bordered"]/tbody/tr/td')[i]
        #for b in prob_w:
        Item_Form.append(item_form.text)  
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Item_Form.append("-")

#Rating
    try:
        rating=driver.find_element(By.XPATH,'//div[@class="bottom-line-items"]/span')
        #for b in price:
        Rating.append(rating.text + " out of 5")  
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Rating.append("-")    

#No_of_Ratings
    try:
        num=driver.find_element(By.XPATH,'//span[@class="reviews-qa-labels-container "]/span')
        #for b in price:
        No_of_Ratings.append(num.text)  
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        No_of_Ratings.append("-")           
        
#IMG_URL
    try:
#         img=driver.find_element_by_xpath('//img[@class="w-100"]')
        
#         class="w-100 lazyloaded"
#         #for b in img:
#         IMG_URl.append(img.get_attribute('src'))
        img=driver.find_elements(By.XPATH,'//img[@class="w-100"]')
        img1 = driver.find_elements(By.XPATH,'//img[@class="w-100 lazyloaded"]')
        if len(img) >= 1:
            img=driver.find_element(By.XPATH,'//img[@class="w-100"]')
            IMG_URl.append(img.get_attribute('src'))
        elif len(img1) >= 1:
                img1=driver.find_element(By.XPATH,'//img[@class="w-100 lazyloaded"]')
                IMG_URl.append(img1.get_attribute('src'))
        else:    
            IMG_URl.append("-")
        
    except NoSuchElementException as e:
        IMG_URl.append("-")
        
    Source.append("Leslies")
    Extract_Date.append(pd.Timestamp.today().strftime('%Y-%m-%d'))    
```

```python
for url in url_top[200:400]:
    time.sleep(3)
    driver.get(url)
    WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//span[@id="productTitle"]')))

#Brand Name
    try:
        brand=driver.find_element(By.XPATH,'//*[@id="bylineInfo"]')
        #for b in brand:
        try:
            Brand_Name_AZ.append(brand.text.split(':')[1].strip())
        except:
            Brand_Name_AZ.append(brand.text)

    except NoSuchElementException as e:
        Brand_Name_AZ.append("-")
        
#ProdName                                            
    try:
        pname=driver.find_elements(By.XPATH,'//span[@id="productTitle"]')
        for b in pname:
            ProdName_AZ.append(b.text.strip())
        
    except NoSuchElementException as e:
        ProdName_AZ.append("-")    
#Price
    try:
        price=driver.find_elements(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayPadding priceToPay"]')
        price1 =driver.find_elements(By.XPATH,'//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
        price2 =driver.find_elements(By.XPATH,'//span[@class="a-size-base a-color-price"]')
        price3 =driver.find_elements(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        if len(price)>=1:
            price=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayPadding priceToPay"]')
            #for b in price:
            Price_AZ.append(price.text.split()[0]+"."+ price.text.split()[1])
        
        elif len(price1)>=1:
            price1 =driver.find_element(By.XPATH,'//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
            #for b in price:
            Price_AZ.append(price1.text)  
        elif len(price3)>=1:
            price3 =driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
            #for b in price:
            Price_AZ.append(price3.text.replace('\n','.'))
       
        else:
            Price_AZ.append("-")                
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Price_AZ.append("-")

#Produt Type
    try:
        prod= driver.find_elements(By.XPATH,'//ul[@class="a-unordered-list a-horizontal a-size-small"]/li/span/a')
        #for b in prob:
        if len(prod) >= 1:
            Produt_Type_AZ.append(prod[-1].text)
        else:
            Produt_Type_AZ.append("-")
            
    except NoSuchElementException as e:
        Produt_Type_AZ.append("-")
       

    
#Manufacturer                                           
    try:
        ld1 = driver.find_elements(By.XPATH,'//ul[@class="a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"]/li')
        d1 = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')
        ls1 = driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/th')
        if len(d1)>=1:
            for i in range(0,len(d1)):
                if driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')[i].text =='Manufacturer':
                    manf = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/td')[i]

        elif len(ld1)>=1:
            for i in range(0,len(ld1)):
                if driver.find_elements(By.XPATH,'//ul[@class="a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"]/li/span/span[1]')[i].text =='Manufacturer :':
                    manf = driver.find_elements(By.XPATH,'//ul[@class="a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"]/li/span/span[2]')[i]
        elif len(ls1)>=1:
            for i in range(0,len(ls1)):
                if driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/th')[i].text =='Brand Name':
                    manf = driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/td')[i]
                    #for b in varient:
        Manufacturer_AZ.append(manf.text)
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Manufacturer_AZ.append("-")
                     
#Product Weight
    try:
        d = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')
        ls = driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/th')
        if len(d)>=1:
            for i in range(0,len(d)):
                if driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')[i].text =='Item Weight':
                    prob_w = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/td')[i]

        elif len(ls)>=1:
            for i in range(0,len(ls)):
                if driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/th')[i].text =='Item Weight':
                    prob_w = driver.find_elements(By.XPATH,'//table[@id="product-specification-table"]/tbody/tr/td')[i]
                #for b in varient:
        Produt_Weight_AZ.append(prob_w.text)        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Produt_Weight_AZ.append("-")

            
#img
    try:
        
        img=driver.find_elements(By.XPATH,'//div[@id="imgTagWrapperId"]/img')
        img1=driver.find_elements(By.XPATH,'//span[@id="a-autoid-3-announce"]/img')
        if len(img)>=1:
            img=driver.find_element(By.XPATH,'//div[@id="imgTagWrapperId"]/img')
            #for b in img:
            IMG_URl_AZ.append(img.get_attribute('src'))
        
        elif len(img1)>=1:
            img1=driver.find_element(By.XPATH,'//span[@id="a-autoid-3-announce"]/img')
            #for b in img:
            IMG_URl_AZ.append(img1.get_attribute('src'))

    except (NoSuchElementException, StaleElementReferenceException) as e:
        IMG_URl_AZ.append("-")
#Best Seller 
    try:
        driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')
        s1 = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')
        if len(s1)>=1:
            for i in range(0,len(s1)):
                if driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/th')[i].text =='Best Sellers Rank':
                    seller1 = driver.find_elements(By.XPATH,'//table[@class="a-keyvalue prodDetTable"]/tbody/tr/td')[i]
            Best_Seller_Rank_AZ.append(seller1.text)
        else:
            Best_Seller_Rank_AZ.append("-") 
            
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Best_Seller_Rank_AZ.append("-")
#Item Form 
    try: 
        form = driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[3]/td[2]/span')
        #for b in form:
        Item_Form_AZ.append(form.text)
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Item_Form_AZ.append("-")
        
#Rating
    try:
        rate=driver.find_element(By.XPATH,'//span[@id="acrCustomerReviewText"]')
        No_of_Ratings.append(rate.text)        # No of Ratings 
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//span[@id="acrCustomerReviewText"]'))).click()
        rating=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-base"]')
        Rating_AZ.append(rating.text)     # Collecting Ratings
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Rating_AZ.append("-")
        No_of_Ratings.append("-")
        
    try:
        stars = driver.find_element(By.XPATH,'//span[@id="acrCustomerReviewText"]')
        if driver.find_element(By.XPATH,'//span[@id="acrCustomerReviewText"]').text != None:
            star_5 = driver.find_element(By.XPATH,'//table[@class="a-normal a-align-center a-spacing-base"]/tbody/tr[1]/td[3]/span[2]')
            star_4 = driver.find_element(By.XPATH,'//table[@class="a-normal a-align-center a-spacing-base"]/tbody/tr[2]/td[3]/span[2]')
            star_3 = driver.find_element(By.XPATH,'//table[@class="a-normal a-align-center a-spacing-base"]/tbody/tr[3]/td[3]/span[2]')
            star_2 = driver.find_element(By.XPATH,'//table[@class="a-normal a-align-center a-spacing-base"]/tbody/tr[4]/td[3]/span[2]')
            star_1 = driver.find_element(By.XPATH,'//table[@class="a-normal a-align-center a-spacing-base"]/tbody/tr[5]/td[3]/span[2]')
            stars_5.append(star_5.text)
            stars_4.append(star_4.text)
            stars_3.append(star_3.text)
            stars_2.append(star_2.text)
            stars_1.append(star_1.text)

        else:
            stars_5.append("-")
            stars_4.append("-")
            stars_3.append("-")
            stars_2.append("-")
            stars_1.append("-")
    except (NoSuchElementException, StaleElementReferenceException) as e:
            stars_5.append("-")
            stars_4.append("-")
            stars_3.append("-")
            stars_2.append("-")
            stars_1.append("-")

    Source_AZ.append("Amazon")
    Extract_Date.append(pd.Timestamp.today().strftime('%Y-%m-%d')) 
```
