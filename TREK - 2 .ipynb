```python
import numpy as np 
import pandas as pd
from datetime import datetime
import os
import warnings
warnings.filterwarnings("ignore")

df=pd.read_excel(r'C:\Users\A803377\Downloads\Badge-Earned-Report - 2024-04-18T122452.274.xlsx')

df2= df.copy()
df2.columns = df.columns.str.partition(':').droplevel(2)
df2.columns = df2.columns.droplevel(1)
df2.columns = df2.columns.str.replace("."," ")

cols=["Group","SubGroup","State/Province","Country"]
for col in cols:
    df2[col] = df2[col].str.replace("," ," ")

df2 = df2.set_index(["First Name","Last Name","Username","Group","SubGroup","State/Province","Country"])

# Boolean List of B,S,G Columns 
B_List = df2.columns.str.endswith('B')
S_List = df2.columns.str.endswith('S')
G_List = df2.columns.str.endswith('G')

# Converting into list to form a Dataframe 
df_B = df2.columns[B_List]
df_S = df2.columns[S_List]
df_G = df2.columns[G_List]

# Useing List to make dataframe
df_B=df2[df_B]
df_S=df2[df_S]
df_G=df2[df_G]

# Replacing vales with "1" For B, "2" for S and "3" for G
df_B=df_B.replace(['Yes'],'3')
df_B=df_B.replace(['yes'],'3')

df_S=df_S.replace(['Yes'],'2')
df_S=df_S.replace(['yes'],'2')

df_G=df_G.replace(['Yes'],'1')
df_G=df_G.replace(['yes'],'1')


Mergelist=["First Name","Last Name","Username","Group","SubGroup","State/Province","Country"]

Column_format = list(df2.columns)
df_Final = pd.merge(pd.merge(df_B, df_S,on= Mergelist, how="left"),df_G,on= Mergelist,how="left")
df_Final = df_Final.reindex(Column_format, axis=1)

df_Final = df_Final.replace(['No','no'], np.NaN)

df_Final.reset_index(inplace = True)

file_name = 'FinalTREK_' + str(datetime.now().strftime('%Y-%m-%d')) + '.csv'

df_Final.to_csv(file_name,index=False)
```
