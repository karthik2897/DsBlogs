```python
for url in URL_THD[130:160]:
    time.sleep(3)
    driver.get(url)
#     time.sleep(2)
    value = 0
    for i in range(5):
        driver.execute_script("scrollBy("+ str(value) +",+1400);")
        value += 1400
        time.sleep(2)

#     WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//div[@class="reviews-histogram"]')))

        
#ProdName                                            
    try:
        pname=driver.find_element(By.XPATH,'//div[@class="title"]')
        ProdName_THD.append(pname.text.strip())
        
    except NoSuchElementException as e:
        ProdName_THD.append("-")    
#Price
    try:
        
        driver.find_element(By.XPATH,'//div[@class="price-format__large price-format__main-price"]/span')
        price=driver.find_elements(By.XPATH,'//div[@class="price-format__large price-format__main-price"]/span')[1]
        price1=driver.find_elements(By.XPATH,'//div[@class="price-format__large price-format__main-price"]/span')[2]
        Price_THD.append('$'+price.text +"."+ price1.text)
        
    except NoSuchElementException as e:
        price2= driver.find_element(By.XPATH,'//span[@class="price-detailed__unit-price"]/span')
        Price_THD.append(price2.text)
        
    except :
        Price_THD.append("-")


##Manufacturer                                           
    try:
        varient=driver.find_element(By.XPATH,'//h2[@class="product-details__brand--link"]')
        Manufacturer_THD.append(varient.text)  
        
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Manufacturer_THD.append("-")

    try:
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//div[@class="sticky-nav__jump-links"]/ul/li[2]'))).click()
        d = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//table[@class="table-container__table--8up88"]/tbody/tr/th')))
        #Produt Type
        try:
            d = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')
            for i in range(0,len(d)):
                if driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')[i].text =='Outdoor Living Product Type':
                    prob = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/td')[i]
            Produt_Type_THD.append(prob.text)  

        except (NoSuchElementException, StaleElementReferenceException) as e:
            Produt_Type_THD.append("-")

        #Product Weight
        try:
            d = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')
            for i in range(0,len(d)):
                if driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')[i].text =='Product Weight (lb.)':
                    prob_w = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/td')[i]
            Produt_Weight_THD.append(prob_w.text)  

        except (NoSuchElementException, StaleElementReferenceException) as e:
            Produt_Weight_THD.append("-")


        #Item_Form
        try:
            d = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')        
            for i in range(0,len(d)):
                if driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/th')[i].text =='Pool Chemical Type':
                    item_form = driver.find_elements(By.XPATH,'//table[@class="table-container__table--8up88"]/tbody/tr/td')[i]
            Item_Form_THD.append(item_form.text)  

        except (NoSuchElementException, StaleElementReferenceException) as e:
            Item_Form_THD.append("-")
            
    except (NoSuchElementException, StaleElementReferenceException) as e:
        Produt_Type_THD.append("-")
        Produt_Weight_THD.append("-")
        Item_Form_THD.append("-")
    value = 0
    for i in range(2):
        driver.execute_script("scrollBy("+ str(value) +",+1400);")
        value += 1400
        time.sleep(2)
    try:
        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, '//div[@class="navlink-rr"]'))).click()
        time.sleep(2)
        #     WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//span[@class="ratings-and-reviews-overview__desktop__list__review-average"]')))
        #Rating
        try:
            rating=driver.find_element(By.XPATH,'//span[@class="ratings-and-reviews-overview__desktop__list__review-average"]')
            #for b in price:
            Rating_THD.append(rating.text)  

        except (NoSuchElementException, StaleElementReferenceException) as e:
            Rating_THD.append("-")    

        #No_of_Ratings
        try:
            num=driver.find_element(By.XPATH,'//span[@class="product-details__review-count"]')
            #for b in price:
            No_of_Ratings.append(num.text.strip("()"))  

        except (NoSuchElementException, StaleElementReferenceException) as e:
            No_of_Ratings.append("-")           

    except (NoSuchElementException, StaleElementReferenceException) as e:
        Rating_THD.append("-") 
        No_of_Ratings.append("-")
        
#IMG_URL
    try:
        img=driver.find_element(By.XPATH,'//button[@class="mediagallery__mainimage--clickable"]/div/img')
        #for b in img:
        IMG_URl_THD.append(img.get_attribute('src'))

    except NoSuchElementException as e:
        IMG_URl_THD.append("-")
#Brand Name    
    try:
        driver.find_element(By.XPATH,'//span[@class="product-details__brand--link"]')
        brand=driver.find_element(By.XPATH,'//span[@class="product-details__brand--link"]')
        Brand_Name_THD.append(brand.text.strip())
    except NoSuchElementException as e:
        brand=driver.find_element(By.XPATH,'//div[@class="sticky-nav__brand--container"]')
        Brand_Name_THD.append(brand.text.strip())

    except NoSuchElementException as e:
        Brand_Name_THD.append("-")        

#Rating

    try:
        stars = driver.find_element(By.XPATH,'//div[@class="reviews-histogram"]')
        if driver.find_element(By.XPATH,'//div[@class="reviews-histogram"]').text != None:
            star_5 = driver.find_elements(By.XPATH,'//div[@class="reviews-histogram"]/div')[0]
            star_4 = driver.find_elements(By.XPATH,'//div[@class="reviews-histogram"]/div')[1]
            star_3 = driver.find_elements(By.XPATH,'//div[@class="reviews-histogram"]/div')[2]
            star_2 = driver.find_elements(By.XPATH,'//div[@class="reviews-histogram"]/div')[3]
            star_1 = driver.find_elements(By.XPATH,'//div[@class="reviews-histogram"]/div')[4]

            stars_5.append(star_5.text.split()[1])
            stars_4.append(star_4.text.split()[1])
            stars_3.append(star_3.text.split()[1])
            stars_2.append(star_2.text.split()[1])
            stars_1.append(star_1.text.split()[1])

        else:
            stars_5.append("-")
            stars_4.append("-")
            stars_3.append("-")
            stars_2.append("-")
            stars_1.append("-")
    except (NoSuchElementException, StaleElementReferenceException) as e:
            stars_5.append("-")
            stars_4.append("-")
            stars_3.append("-")
            stars_2.append("-")
            stars_1.append("-")


    Source_THD.append("The Home Depot")
    Extract_Date.append(pd.Timestamp.today().strftime('%Y-%m-%d'))

```
