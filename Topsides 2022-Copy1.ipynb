```python
df_oneStream = df_oneStream[[' Entity',' Account',' UD1',' Amount',' Time']]
df_oneStream.rename(columns = {' Entity':'Company Code',' UD1': 'Reporting_Line',' Account':'Account',' Amount':'Amount',' Time':'Fiscal Period'}, inplace = True)
df_oneStream= df_oneStream.groupby(["Company Code","Account","Reporting_Line",'Fiscal Period'],as_index=False)["Amount"].sum()
df_oneStream = df_oneStream.pivot_table(index=["Company Code","Reporting_Line",'Fiscal Period'],columns='Account',values='Amount',aggfunc='sum',).fillna(0)
df_oneStream.rename(columns = {'MGMT_GR_Prof':'GP'}, inplace = True)
df_oneStream.reset_index(inplace=True)

# Changes
df_oneStream['Company Code'] = df_oneStream['Company Code'].apply(lambda x: '{0:0>4}'.format(x))
df_oneStream["Reporting_Line"]= df_oneStream["Reporting_Line"].replace({'LAM_IWT':'LAM_IS', 'LAM_PP':'LAM_CS','APAC_PP':'AP_CS','APAC_IWT':'AP_IS',
                                                                    'EMEA_PP':'EMEA_CS','EMEA_IWT':'EMEA_IS','NA_PP':'NA_CS','NA_IWT':'NA_IS',
                                                                   'PS_EMEA':'EMEA_PS','PS_LA':'SA_PS','PS_NAM':'NA_PS','PS_SA':'ZA_PS'},regex=True)

df_oneStream["Reporting_Line"]= df_oneStream["Reporting_Line"].replace('None', 'NVOther',regex=True)
df_oneStream["Reporting_Line"]= df_oneStream["Reporting_Line"].replace('No_SG', 'NVOther',regex=True)
df_oneStream = df_oneStream.groupby(['Company Code','Reporting_Line','Fiscal Period'],as_index=False).sum()

df_oneStream
```

```python
df_oneStream["Fiscal Period"]= df_oneStream["Fiscal Period"].replace({'2023M1': '2023001','2023M2': '2023002','2023M3': '2023003','2023M4': '2023004','2023M5': '2023005','2023M6': '2023006','2023M7': '2023007','2023M8': '2023008','2023M9': '2023009'},regex=True)
df_oneStream["Fiscal Period"] = df_oneStream["Fiscal Period"].astype('string')


# df_oneStream["Fiscal Period"]= df_oneStream["Fiscal Period"].replace({'2023M6': '2023006'},regex=True)
# df_oneStream["Fiscal Period"] = df_oneStream["Fiscal Period"].astype('string')
```
