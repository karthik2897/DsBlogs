```python
import win32com.client
import os
from datetime import datetime, timedelta
import pytz  # Import pytz for timezone handling

# Connect to Outlook
outlook = win32com.client.Dispatch("Outlook.Application")
namespace = outlook.GetNamespace("MAPI")

# Access the shared mailbox
email_id = 'automationpx-team@solenis.com'  # Replace with the actual email ID
shared_mailbox = namespace.Folders[email_id]

# Access the Inbox of the shared mailbox
inbox = shared_mailbox.Folders['Inbox']  # Access Inbox

# Directory to save attachments
attachments_dir = r'C:\Users\A803377\Emails'  # Replace with your local path
os.makedirs(attachments_dir, exist_ok=True)

# Specify the date range (e.g., last 10 days)
days_to_check = 33  # Change to 20 for the last 20 days
start_date = datetime.now(pytz.utc) - timedelta(days=days_to_check)  # Make start_date timezone-aware

# Specify the sender's email address to filter by
# sender_email = 'reports-noreply@five9.com'  # Replace with the actual sender's email
sender_email = 'I.T@zhgplc.com'  # Replace with the actual sender's email



# Collect emails
emails = inbox.Items
emails.Sort("[ReceivedTime]", True)  # Sort emails by received time (most recent first)

# Loop through emails and process those within the specified date range and sender
for item in emails:
    # Check if the email is from the specified sender
    if item.SenderEmailAddress.lower() != sender_email.lower():
        continue  # Skip emails not from the specified sender

    # Get the received time in UTC
    received_time = item.ReceivedTime  # This should already be in the correct format

    # Check if the email is from the specified sender and within the date range
    if received_time < start_date or item.SenderEmailAddress.lower() != sender_email.lower():
        continue  # Skip emails outside the date range or not from the specified sender

    try:
        print(f"Processing email: {item.Subject}")

        if item.Attachments.Count > 0:
            # Use the received date of the email for the report date
            report_date = received_time.strftime("%Y-%m-%d")

            for attachment in item.Attachments:
                attachment_name = attachment.FileName
                attachment_path = os.path.join(attachments_dir, attachment_name)

#                 Logic for renaming files based on email subject
                if "EU Domain" in item.Subject:
                    new_file_name = f"{report_date} - Daily Calls - Five9 EU Domain.csv"
                    attachment_path = os.path.join(attachments_dir, new_file_name)

                if "US Domain" in item.Subject:
                    new_file_name = f"{report_date} - Daily Calls - Five9 US Domain.csv"
                    attachment_path = os.path.join(attachments_dir, new_file_name)
                    
                if "Order Line" in item.Subject:
                    new_file_name = f"{report_date} - Daily Orders - Zenith - Order Lines.csv"
                    attachment_path = os.path.join(attachments_dir, new_file_name)

                if "Order Header" in item.Subject:
                    new_file_name = f"{report_date} - Daily Orders - Zenith - Order Headers.csv"
                    attachment_path = os.path.join(attachments_dir, new_file_name)
                # Save the attachment
                attachment.SaveAsFile(attachment_path)
                print(f"Saved attachment: {new_file_name} from email: {item.Subject}")
        else:
            print(f"No attachments in email: {item.Subject}")

    except Exception as e:
        print(f"Error processing email: {e}")

print(f"Attachments downloaded to {attachments_dir}")
```
