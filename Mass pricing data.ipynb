```python
df_Amazon=pd.read_excel(r'C:\Users\A803377\Amazon1 -2022-12-28.xlsx')
df_leslies=pd.read_excel(r'C:\Users\A803377\Leslies1 -2022-12-28.xlsx')
df_Walmart=pd.read_excel(r'C:\Users\A803377\Walmart -2022-12-29.xlsx')
df_THD=pd.read_excel(r'C:\Users\A803377\TheHomeDepot1 -2022-12-28.xlsx')
```

```python
df_Walmart.rename(columns = {'IMG_URL':'IMG_URl'}, inplace = True)
```

```python
df_THD
```

```python
df_THD.Produt_Weight = df_THD.Produt_Weight.str.replace('[^0-9.]', '')
df_THD['Produt_Weight'] = df_THD['Produt_Weight'] + ' pounds'
```

```python
df_Amazon = df_Amazon.replace("-",np.NaN)
df_Amazon = df_Amazon.replace('%','',regex=True)
df_Amazon[['stars_5','stars_4','stars_3','stars_2','stars_1']] = df_Amazon[['stars_5','stars_4','stars_3','stars_2','stars_1']].astype(float)*(1/100)
```

```python
df_Walmart[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1']] = df_Walmart[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1']].astype(str).replace("0",np.NaN)
df_Walmart['stars_5'] = df_Walmart['stars_5'].astype(float)/df_Walmart['No_of_Ratings'].astype(float)
df_Walmart['stars_4'] = df_Walmart['stars_4'].astype(float)/df_Walmart['No_of_Ratings'].astype(float)
df_Walmart['stars_3'] = df_Walmart['stars_3'].astype(float)/df_Walmart['No_of_Ratings'].astype(float)
df_Walmart['stars_2'] = df_Walmart['stars_2'].astype(float)/df_Walmart['No_of_Ratings'].astype(float)
df_Walmart['stars_1'] = df_Walmart['stars_1'].astype(float)/df_Walmart['No_of_Ratings'].astype(float)
df_Walmart[['stars_5','stars_4','stars_3','stars_2','stars_1']] = df_Walmart[['stars_5','stars_4','stars_3','stars_2','stars_1']].round(2)
```

```python
df_THD = df_THD.replace("-",np.NaN)
df_THD[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1']] = df_THD[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1']].astype(str).replace("0",np.NaN)
df_THD['stars_5'] = df_THD['stars_5'].astype(float)/df_THD['No_of_Ratings'].astype(float)
df_THD['stars_4'] = df_THD['stars_4'].astype(float)/df_THD['No_of_Ratings'].astype(float)
df_THD['stars_3'] = df_THD['stars_3'].astype(float)/df_THD['No_of_Ratings'].astype(float)
df_THD['stars_2'] = df_THD['stars_2'].astype(float)/df_THD['No_of_Ratings'].astype(float)
df_THD['stars_1'] = df_THD['stars_1'].astype(float)/df_THD['No_of_Ratings'].astype(float)
df_THD[['stars_5','stars_4','stars_3','stars_2','stars_1']] = df_THD[['stars_5','stars_4','stars_3','stars_2','stars_1']].round(2)
```

```python
df_merged = pd.concat([ df_Amazon,df_leslies,df_Walmart,df_THD], axis=0,ignore_index=True)
```

```python
# Data Cleaning 

df_merged[['Packaging Weight', 'Packaging Weight UOM']] = df_merged['Produt_Weight'] .str.split(' ', expand=True)
df_merged = df_merged.replace(np.NaN,"-")
df_merged = df_merged.replace('nan',"-")
df_merged['Price'] = df_merged['Price'].str.replace("$" ,"")
df_merged['No_of_Ratings'] = df_merged['No_of_Ratings'].str.replace(" ratings" ,"")
df_merged['No_of_Ratings'] = df_merged['No_of_Ratings'].str.replace(" rating" ,"")
df_merged['No_of_Ratings'] = df_merged['No_of_Ratings'].str.replace(" Reviews" ,"")
df_merged['No_of_Ratings'] = df_merged['No_of_Ratings'].str.replace(" Review" ,"")
df_merged['Rating'] = df_merged['Rating'].str.replace(' out of 5','',)
df_merged['Rating'] = df_merged['Rating'].str.replace(' star rating' ,'')
df_merged['Brand_Name'] = df_merged['Brand_Name'].str.replace("Visit the " ,"")
df_merged = df_merged.replace("-",np.NaN)
df_merged.Price = df_merged.Price.str.replace('[^0-9.]', '')

cols=['Price','Rating','No_of_Ratings','Packaging Weight']
for col in cols:
    df_merged[col] = df_merged[col].str.replace("," ,"")
    df_merged[col] = df_merged[col].str.replace(" " ,"")
```

```python

# Changing Data Types to match SQL Datatypes 
df_merged[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1','Packaging Weight']] = df_merged[['Rating','No_of_Ratings','stars_5','stars_4','stars_3','stars_2','stars_1','Packaging Weight']].apply(pd.to_numeric,errors='coerce')
df_merged[["Price"]] = df_merged[["Price"]].apply(pd.to_numeric)
df_merged[["Extract_Date"]] = df_merged[["Extract_Date"]].apply(pd.to_datetime)

# Final Table 
df_merged.index = np.arange(1, len(df_merged)+1)
df_merged.rename_axis('ID', inplace=True)
df_merged.reset_index(level=0, inplace=True)
```
